## Below function returns df of video title, views, uploaded which time ago, links of videos

Input of the function is youtube channel videos page link

```python
def helper(link):
    driver=webdriver.Chrome(r"C:\Users\pk\Downloads\chromedriver_win32\chromedriver.exe")
    driver.get('https://www.'+link)

    prev_h = 0
    n=3
    while n:
        height = driver.execute_script("""
                function getActualHeight() {
                    return Math.max(
                        Math.max(document.body.scrollHeight, document.documentElement.scrollHeight),
                        Math.max(document.body.offsetHeight, document.documentElement.offsetHeight),
                        Math.max(document.body.clientHeight, document.documentElement.clientHeight)
                    );
                }
                return getActualHeight();
            """)
        driver.execute_script(f"window.scrollTo({prev_h},{prev_h + 200})")
        time.sleep(1)
        prev_h +=200  
        if prev_h >= height:
            break
        n-=1
        
    content=driver.page_source
    soup=BeautifulSoup(content,'html.parser')

    title=soup.select_one('#container h1').text
    date=soup.select('#info-strings yt-formatted-string')[0].text
    views=soup.find('span', class_="view-count style-scope ytd-video-view-count-renderer").text
    likes=soup.find('yt-formatted-string',id='text',class_='style-scope ytd-toggle-button-renderer style-text').get('aria-label')
    comments=soup.select_one(".ytd-comments-header-renderer .ytd-comments-header-renderer .yt-formatted-string").text
    duration = soup.find("span", {"class": "ytp-time-duration"}).text
    return title,date,views,likes,comments,duration
```

```python
def channel(link):
    driver=webdriver.Chrome(r"C:\Users\pk\Downloads\chromedriver_win32\chromedriver.exe")
    driver.get(link)

    prev_h = 0
    while True:
        height = driver.execute_script("""
                function getActualHeight() {
                    return Math.max(
                        Math.max(document.body.scrollHeight, document.documentElement.scrollHeight),
                        Math.max(document.body.offsetHeight, document.documentElement.offsetHeight),
                        Math.max(document.body.clientHeight, document.documentElement.clientHeight)
                    );
                }
                return getActualHeight();
            """)
        driver.execute_script(f"window.scrollTo({prev_h},{prev_h + 300})")
        time.sleep(1)
        prev_h +=300  
        if prev_h >= height:
            break

    content=driver.page_source.encode('utf-8').strip()
    soup=BeautifulSoup(content,'html.parser')
    titles=[i.text for i in soup.find_all('a',id='video-title')]
    views_time=[i.text for i in soup.find_all('span',class_='style-scope ytd-grid-video-renderer')]
    views=[]
    times=[]
    for i in range(0,len(views_time),2):
        views.append(views_time[i].split()[0])
        times.append(views_time[i+1])
    urls=['youtube.com'+i.get('href') for i in soup.find_all('a',id='video-title')]
    df=pd.DataFrame({'name':titles,'views':views,'uploaded':times,'link':urls})
    df.to_csv('output.csv',index=False)
    return df
```
